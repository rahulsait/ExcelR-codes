## SEABORN :

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading an inbuilt dataset - tips :
tips = sns.load_dataset("tips")
```

```python
tips
```

### Univariate - one variable data visualization :

```python
# Using strip plot :
sns.stripplot(y = "tip", data = tips, jitter = True)
plt.ylabel("tip $")
plt.show()
```

```python
# use of jitter :
sns.stripplot(y = "tip", data = tips, jitter = False)
plt.ylabel("tip ($)")
plt.show()
```

so if jitter = False, we get a single dense line which is not much of a use to mKe inferences.

we can see that the most dense area is between y = 2 to 4
which indicates that most tips given is between 2$ and 4$

```python
# grouping with stripplot :
sns.stripplot(x = "day", y = "tip", data = tips)
plt.ylabel("tip ($)")
plt.show()
```

we can infer from the above plot is that most tips were given on saturday

```python
sns.stripplot(x = "sex", y = "tip", data = tips)
plt.ylabel("tip ($)")
plt.show()
```

most tips were given by male than by female

```python
# Using swarm plot to infer more from the data than stripplot :
sns.swarmplot(x = "day", y = "tip", data = tips)
plt.ylabel("tip ($)")
plt.show()
```

```python
# adding another parameter to infer some more : based on sex
sns.swarmplot(x = "day", y = "tip", data = tips, hue(color) = "sex")
plt.ylabel("tip ($)")
plt.show()
```

```python
# making the plot horizontal :
sns.swarmplot(x = "day", y = "tip", data = tips, hue = "sex", orient = "h")
plt.ylabel("tip ($)")
plt.show()
```

```python
# Using Box Plot to infer :
sns.boxplot(x = "day", y ="tip", data = tips)
plt.ylabel("tip ($)")
plt.show()
```

we can infer from the above boxplot that :
on sat tips were given in the range 6 - 10 also 
on sun the average tips among all days was the highest

```python
# using violin plot :
sns.violinplot(x = "day", y = "tip", data = tips)
plt.ylabel("tip ($)")
plt.show()
```

```python
# combining plots :
plt.subplot(1,2,1) # 1,2 - size and 1 - position
sns.boxplot(x = "day", y = "tip", data = tips)
plt.ylabel("tip ($)")
plt.subplot(1,2,2) # 1,2 - size and 2 - position
sns.violinplot(x = "day", y = "tip", data = tips)
plt.ylabel("tip ($)")
plt.show()
```

```python
# not giving positions of the plots :
sns.violinplot(x = "day", y = "tip", data = tips, inner = None, color = "silver")
sns.stripplot(x = "day", y = "tip", data = tips, size = 4, jitter = True)
plt.ylabel("tip ($)")
plt.show()
```

## Bivariate - two variable data visualization :

```python
# Using Joint plot to plot between two or more variables :
sns.jointplot(x = "total_bill", y = "tip", data = tips)
plt.show()
```

from the above joint plot we can infer that the tips given are most when the total bill is between 10 to 20 $

```python
# Plotting the same plot by using kde (kernel density estimation)
# kde is a non parametric way to estimate the probability density function a random variable
sns.jointplot(x = "total_bill", y = "tip", data = tips, kind = "kde")
plt.show()
```

The first incircle in the middle of the plot indicates the most denser part - 
on x axis : between 10 - 18
on y axis : between 1.8 - 2.8

```python
# using pair plot to get histograms as well as scatterplots between all the variables of a dataset :
sns.pairplot(tips)
plt.show()
```

In the above pair plot we usually consider the lower triangle plots

```python
sns.pairplot(tips, hue = "sex")
plt.show()
```

## Titanic Dataset :

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
titanic = pd.read_csv("/content/train (1).csv")
```

```python
titanic.head()
```

```python
titanic
```

```python
titanic.columns
```

```python
# let's check the data types first
titanic.dtypes
```

The Target variable column is "Survived" in this data set.

```python
# making inferences on the survivals based on the passenger ID by using groupy function ( for categorical variables) :
titanic.groupby("Survived")["PassengerId"].count()
```

```python
# we infer that 342 passengers were rescued and survived and 549 passengers did not survive
```

```python
f, ax = plt.subplots(figsize = (11,5))
sns.boxplot(x = "Survived", y = "Age", data = titanic)
plt.show()
```

```python
# we can infer from the above box plot that the "age" was not a significant feature for the rescuing operations
```

```python
# comparing the survivals with sex :
f, ax = plt.subplots(figsize = (11,7))
sns.boxplot(x = "Sex", y = "Survived", data = titanic)
```

```python
  # adding age parameter too :
  f, ax = plt.subplots(figsize = (8,6))
  sns.boxplot(x = "Sex", y = "Age", hue = "Survived", data = titanic)
  plt.show()
```

```python
# using barplot (best used for discrete variables) to compare between the Passenger classes and their survivals :
sns.barplot(x = "Pclass", y = "Survived", data = titanic)
plt.show()
```

```python
# from the above barplot we can infer that the passengers in higher classes were rescued the most than the lower ones.
# so the rescue operations have been prioritized based on the passenger class
```

```python
# lets check wrt sex along with the above variables :
sns.barplot(x = "Pclass", y = "Survived", hue = "Sex", data = titanic)
plt.show()
```

```python
# we can tell that the females were given a greater priority in every class than men were given.
```

```python
# comparing SibSp (discrete) with Survivals :
sns.barplot(x = "SibSp", y = "Survived", data = titanic)
plt.show()
```

```python
# passengers having 1 sibling/spouse were rescued the most as compared to others
```

```python
sns.barplot(x = "SibSp", y = "Survived", hue = "Sex", data = titanic)
plt.show()
```

```python
# among all the passengers - female passengers having no sibling/spouse were rescued the most.
```

```python
# comparing Parch (Parents or children) of passengers with the survival rate :
sns.barplot(x = "Parch", y = "Survived", data = titanic)
plt.show()
```

```python
# no particular patter observed here
```

lets plot a distribution plot of Age wrt the passengers who survived :

```python
survived = titanic.loc[titanic["Survived"] == 1, "Age"].dropna()
sns.distplot(survived)
plt.title("Survived")
plt.plot()
```

```python
# from the above histogram we can tell that the passengers with age 20 - 35 were survived the most
```

```python
# dist plot for non - survivals :
not_survived = titanic.loc[titanic["Survived"] == 0, "Age"]
sns.distplot(not_survived)
plt.title("Not Survived")
plt.show()
```

```python
# same as survived count - passengers between 20 - 35 have not been survived
```

```python
# plotting among all the variables - bar plots and scatter plots : using PairPlot
sns.pairplot(titanic.dropna())
plt.plot()
```

```python
sns.pairplot(titanic.dropna(), hue = "Sex")
plt.plot()
```
