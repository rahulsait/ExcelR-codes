# **Isolation Forest :**

```python
# importing IF libs :

from sklearn.ensemble import IsolationForest
```

```python
data = pd.read_csv("iris.csv", index_col = 0)
data_encoded = pd.get_dummies(data)
```

```python
data_encoded
```

```python
# let's add a row at the end to check whether the point is an outlier or not :

data_encoded.loc[150] = [10, 20, 30, 40, 0, 1, 0]

data_encoded
```

```python
# training the model :

clf = IsolationForest(random_state = 10, contamination = 0.1) # random_state : to keep the index consistent(can be any integer),contamination : % of outliers assumption
clf.fit(data_encoded)
```

```python
# predictions : ('1' - inliers, '-1' - outliers)

y_pred_outliers = clf.predict(data_encoded)

y_pred_outliers
```

```python
data_encoded["scores"] = clf.decision_function(data_encoded)
```

```python
data_encoded["anamoly"] = clf.predict(data_encoded.iloc[:, 0:7])
```
