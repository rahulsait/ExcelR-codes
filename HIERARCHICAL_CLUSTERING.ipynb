```python
# import Hierarchical Clustering Lib :

import scipy.cluster.hierarchy as sch # for plotting the Dendrogram
from sklearn.cluster import AgglomerativeClustering # for generating Clusters
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sn
```

```python
uni = pd.read_csv("/content/Universities.csv")
```

```python
uni.head()
```

```python
# Normalization functions :

def norm_func(i) :
  x = (i-i.min())/(i.max()-i.min())
  return(x)
```

```python
# Normalized Dataframe :

df_norm = norm_func(uni.iloc[:,1:])
```

```python
# Creating Dendrogram :

dendrogram = sch.dendrogram(sch.linkage(df_norm, method = "single"))
```

**from the above plot we can see there are 3 clusters**

```python
# Creating Clusters :

hc = AgglomerativeClustering(n_clusters = 4, affinity = "euclidean", linkage = "single")
```

```python
# Save Clusters for charts :

y_hc = hc.fit_predict(df_norm)
Clusters = pd.DataFrame(y_hc, columns = ["Clusters"])
```

```python
Clusters
```

```python
uni["h_clusterid"] = Clusters
```
