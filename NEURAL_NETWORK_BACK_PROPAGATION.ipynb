```python
# creating first MLP(Multi Layer Perceptron) in Keras :

from keras.models import Sequential
from keras.layers import Dense
import numpy
```

```python
# fix random seed for reproducibility :

seed = 7
numpy.random.seed(seed)
```

```python
# load the dataset :

dataset = numpy.loadtxt("/content/pima-indians-diabetes.data (1).csv", delimiter = ",")

# split into input X and output Y variables :

X = dataset[:,0:8]
Y = dataset[:,8]
```

```python
# creating the model :

model = Sequential()

# adding the first input layer :
model.add(Dense(12, input_dim = 8, activation = "relu")) # 12 neurons and 8 columns

# adding the hidden layer :
model.add(Dense(8, activation = "relu")) # 8 neurons

# adding the output layer :
model.add(Dense(1, activation = "sigmoid")) # 1 neuron
```

```python
# compile the model - model configuration :

model.compile(loss = "binary_crossentropy", optimizer = "adam", metrics = ["accuracy"]) # loss - error function
```

```python
# fit the model - model building :

model.fit(X,Y, validation_split = 0.33, epochs = 250, batch_size = 20)
```

```python
# evaluate the model :

scores = model.evaluate(X, Y)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
