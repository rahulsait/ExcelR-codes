```python
measures = pd.Series([24.23,25.53,25.41,24.14,29.62,28.25,25.81,24.39,40.26,32.95,91.36,25.99,39.42,26.71,35.00 ])
```

```python
Company_name = ["Allied Signal","Bankers Trust","General Mills","ITT Industries","JP Morgan & Co.","Lehman Brothers","Marriott","MCI","Merrill Lynch","Microsoft","Morgan Stanley","Sun Microsystems","Travelers","US Airways","Warner-Lambert"]
```

```python
# Plotting the above data in Pie :
plt.figure(figsize = (8,10))
plt.pie(measures, labels = Company_name, autopct = "%1.1f%%")
plt.show()
```
