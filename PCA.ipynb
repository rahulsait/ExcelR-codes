```python
# Import PCA lib :

import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
```

```python
# Loading the data set :

uni = pd.read_csv("/content/PCA.csv")
uni.head()
```

```python
uni.describe()
```

```python
# Considering only Numeric Data :

uni.data = uni.iloc[:,1:]
uni.data.head()
```

```python
# Converting the data into an array :

UNI = uni.data.values
UNI
```

```python
# Scaling the data by Normalization :

uni_normal = scale(UNI)
uni_normal
```

```python
# Applying PCA on this array :

pca = PCA()
pca_values = pca.fit_transform(uni_normal)
pca_values
```

```python
pca = PCA(n_components = 6)
pca_values = pca.fit_transform(uni_normal)
pca_values
```

```python
# Computing the amount of variance from each PCA :

var = pca.explained_variance_ratio_
var
```

```python
# Computing the Cummulative variance : (variance of 1st comp, combined variance of first 2 comps, combined variance of first three comps.)

var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
# variance plot for PCA components obtained :

plt.plot(var1, color = "red")
```

```python
# PCA values of first component :

pca_values[:,0:1]
```

```python
# PCA values for first 2 components :

pca_values[:,0:2]
```

```python
# Plot between PCA1 and PCA2 :

x = pca_values[:,0:1]
y = pca_values[:,1:2]
plt.scatter(x,y)
```

```python
# assigning University names to the points :

finalDf = pd.concat([pd.DataFrame(pca_values[:,0:2], columns = ["PC1","PC2"]), uni[["Univ"]]], axis  =1)
```
