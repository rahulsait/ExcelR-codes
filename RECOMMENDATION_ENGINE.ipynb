```python
# Importing the lib :

import numpy as np
import pandas as pd
```

```python
# Load the dataset :

movies_df = pd.read_csv("/content/Movie.csv")
```

```python
movies_df.head()
```

```python
# No. of unique users in the dataset :

len(movies_df.userId.unique())
```

```python
# No. of unique movies :

len(movies_df.movie.unique())
```

```python
# Changing the format of the dataset :

user_movies_df = movies_df.pivot(index = "userId", columns = "movie", values = "rating").reset_index(drop = True)
```

```python
user_movies_df
```

```python
# Changing the index - userIds :

user_movies_df.index = movies_df.userId.unique()
```

```python
user_movies_df
```

```python
# Input all the NaN values with 0 :

user_movies_df.fillna(0, inplace = True)
```

```python
user_movies_df
```

```python
# Calculating Cosine Similarity between the users :
# Importing Recommendation Libs :

from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances(user_movies_df.values, metric = "cosine")
```

```python
user_sim
```

```python
# Store the results in a dataframe :

user_sim_df = pd.DataFrame(user_sim)
```

```python
# Set the index and column names to user ids

user_sim_df.index = movies_df.userId.unique()
user_sim_df.columns = movies_df.userId.unique()
```

```python
user_sim_df.iloc[0:5,0:5]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5,0:5]
```

```python
# Most similar users :

user_sim_df.idxmax(axis = 1)[0:5]
```
